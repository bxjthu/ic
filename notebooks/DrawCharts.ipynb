我们在日常研究中，有时会碰到需要求复杂方程解的问题，例如：

$$ sin(x)+\frac{1}{x}+1 = cos(x^2) + log(x) $$

在有多个解的时候，二分法用起来较为麻烦，这个时候，就可以使用Python的matplotlib库来画出函数图像，用来判断解的位置

在这个例子中，我画出了这两个函数在[1,10]内的图像

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
x = np.linspace(1, 10, 1000)
y = np.sin(x)+1/x+1
z = np.cos(x**2)+np.log(x)

plt.figure(figsize=(10,6))
plt.plot(x,y,label="$sin(x)+1/x+1$",color="red",linewidth=2)
plt.plot(x,z,"b--",label="$cos(x^2)+log(x)$")
plt.xlabel("x")
plt.ylabel("y")
plt.title("PyPlot Example")
plt.ylim(-1.2,4)
plt.legend()
plt.show()
```

当然，函数图像不止有求方程组的作用，用matplotlib画出的图像来表达统计数据也非常美观,下面是两个柱状统计图以及一个散点分布图


```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))

plt.barh(y_pos, performance, align='center', alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('Performance')
plt.title('How fast do you want to go today?')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


n_groups = 5

means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)

means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index, means_men, bar_width,
                 alpha=opacity,
                 color='b',
                 yerr=std_men,
                 error_kw=error_config,
                 label='Men')

rects2 = plt.bar(index + bar_width, means_women, bar_width,
                 alpha=opacity,
                 color='r',
                 yerr=std_women,
                 error_kw=error_config,
                 label='Women')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D', 'E'))
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# The slices will be ordered and plotted counter-clockwise.
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')
plt.show()
```
